### Jupyter

```python
import pandas as pd

data = pd.read_csv("grades.csv")
data
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

X = data[['English', 'Algebra']]

linked = linkage(X, method='single', metric='euclidean')

# vykreslení dendrogramu
plt.figure(figsize=(10, 5))

dendrogram(linked,
        labels=data['Student'].tolist(),
        leaf_rotation=90,
        distance_sort='descending',
        show_leaf_counts=True,)

plt.title("Dendrogram - English & Math Grades")
plt.xlabel("Student")
plt.ylabel("Vzdálenost")
plt.show()
```
